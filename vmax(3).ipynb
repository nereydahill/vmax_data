```python
import pandas as pd

def parse_data(filename: str) -> list[pd.DataFrame]:
    all_data = []  # List to store data for each run
    with open(filename, 'r') as file:
        contents = file.read()
        data_runs = contents.split('Interval=')[1:]
        for idx, run in enumerate(data_runs):
            actual_run = run.split('#* EndTake')[0]
            actual_run_lines = actual_run.splitlines()[9:]
            actual_run_data = [[float(y) for y in x.split('\t')] for x in actual_run_lines[:-1]]
            run_df = pd.DataFrame(actual_run_data, columns=['Time', 'Pressure', 'MWP'])
            run_df['Volume'] = run_df['Time'] * 500 / 60  # Convert time from seconds to minutes
            all_data.append(run_df)
            
            # Save each DataFrame to a CSV file
            csv_filename = f'run_{idx + 1}.csv'
            run_df.to_csv(csv_filename, index=False)

    return all_data
```
